### **1. ¿Qué es Machine Learning? Conceptos clave y aplicaciones**

#### **¿Qué es Machine Learning?**
Machine Learning es una rama de la inteligencia artificial que permite a las computadoras aprender patrones a partir de datos. En lugar de seguir instrucciones específicas escritas por humanos, los modelos de Machine Learning encuentran relaciones en los datos para hacer predicciones o tomar decisiones.

##### **Ejemplo:**
Supongamos que tienes una lista de características de casas (tamaño, número de habitaciones, etc.) y los precios de venta. Un modelo de Machine Learning puede aprender de esos datos y predecir el precio de una casa nueva basada en sus características.

---

#### **Conceptos clave:**
- **Modelo:** Es una representación matemática de un problema que aprende patrones en los datos.
- **Entrenamiento:** Proceso donde el modelo analiza los datos y ajusta sus parámetros para encontrar patrones.
- **Predicción:** Resultado generado por el modelo al recibir nuevos datos.
- **Datos:** Información estructurada en tablas o conjuntos que se usa para entrenar y evaluar modelos.

---

#### **Aplicaciones:**
1. **Salud:** Detectar enfermedades en imágenes médicas.
2. **Finanzas:** Prevenir fraudes en tarjetas de crédito.
3. **Recomendaciones:** Sugerir productos o películas en plataformas como Amazon o Netflix.
4. **Automóviles:** Conducir vehículos autónomos (como los de Tesla).

---

### **2. Tipos de Machine Learning**

Hay tres tipos principales de Machine Learning:  

#### **1. Aprendizaje Supervisado**
- El modelo aprende a partir de datos etiquetados (datos de entrada con una respuesta conocida).  
- **Ejemplo:** Predecir el precio de casas (entrada: características de la casa, salida: precio).  

#### **2. Aprendizaje No Supervisado**
- El modelo trabaja con datos no etiquetados y busca patrones.  
- **Ejemplo:** Agrupar clientes según sus hábitos de compra.  

#### **3. Aprendizaje por Refuerzo**
- Un agente toma decisiones y aprende de sus errores mediante recompensas o penalizaciones.  
- **Ejemplo:** Un robot aprendiendo a caminar.

#### **¿Qué es un dataset?**
Un **dataset** es una colección de datos organizada, generalmente en forma de tabla:
- **Filas:** Cada fila representa un caso (por ejemplo, una flor específica).
- **Columnas:** Cada columna es una característica (por ejemplo, largo del pétalo).

#### **Dataset Iris**
El dataset **Iris** es un conjunto de datos que contiene información sobre tres tipos de flores (**setosa**, **versicolor**, **virginica**) y sus características:

````python
# Importar bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Cargar dataset Iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Exploración inicial
print("Información del dataset:")
print(df.info())

print("\nEstadísticas descriptivas:")
print(df.describe())

# Visualización
sns.countplot(x='species', data=df)
plt.title('Distribución de especies')
plt.show()

sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=df)
plt.title('Relación entre largo y ancho del sépalo')
plt.show()

# Ejercicios
print("Promedio del largo del pétalo:", df['petal length (cm)'].mean())
print("Flores con pétalos mayores a 4 cm:")
print(df[df['petal length (cm)'] > 4])

plt.hist(df['sepal width (cm)'], bins=10, color='green')
plt.title('Histograma del ancho del sépalo')
plt.xlabel('Ancho del sépalo (cm)')
plt.ylabel('Frecuencia')
plt.show()
````
